```python
stns_data = pd.read_csv(stns, encoding="iso-8859-1")
wther_data = pd.read_csv(wther,encoding="iso-8859-1")
stns_df = pd.DataFrame(stns_data)
wther_df = pd.DataFrame(wther_data)
```

```python
#Weather data summary
wther_df.head()
wther_df.count()
wther_df.describe()
```

```python
#Count NAN values
temp_wther = wther_df
temp_wther['Count_NaNs'] = temp_wther['prcp'].isnull()
temp_wther['Count'] = 1
temp_wther = pd.pivot_table(temp_wther,index=["station"], values=['Count','Count_NaNs'],aggfunc=np.sum)
temp_wther['percentage'] = temp_wther['Count_NaNs']/temp_wther['Count']*100
temp_wther
```

```python
#Pivot table to see the significance of nan values  in weather data
wther_df_nan = pd.pivot_table(wther_df, index=['station'], values=['prcp','tobs'])
wther_df_nan
```

```python
stns_df1 = stns_df['station']
stns_df2 = wther_df['station'].unique()
stns_df1 == stns_df2
```

```python
#Replacing NAN values wiith 0 and saving data in clean_hawaii_stns.csv, clean_hawaii_measurements.csv files
clean_stns_df=pd.DataFrame(stns_df)
clean_wther_df=pd.DataFrame(wther_df)
clean_wther_df.drop('Count_NaNs', axis=1, inplace=True)
clean_wther_df.drop('Count', axis=1, inplace=True)
clean_wther_df = clean_wther_df.fillna(0)
clean_stns_df.to_csv("clean_hawaii_stns.csv", index = False)
clean_wther_df.to_csv("clean_hawaii_measurements.csv", index = False)


```
